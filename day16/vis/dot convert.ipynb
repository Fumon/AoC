```python
import re

lines = None
with open('../input', 'r') as f:
    lines = f.readlines()
```

```python
for te in lines:
    m = re.match(r"Valve ([A-Z]{2}).+valve(?:s)? ((?:(?:[A-Z]{2}(?:, |$)))*)", te).groups()
    start = m[0]
    for end in re.findall(r"([A-Z]{2})", m[1]):
        print(f"{start} -- {end}")
```
